```python
#IMPORTING LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
sb.set_style('whitegrid')
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('Onyx Data -DataDNA Dataset Challenge - Top 1000 YouTubers - November 2023.csv',encoding='unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Longitude','Latitude','Gross tertiary education enrollment (%)','created_date','channel_type_rank','country_rank','video_views_rank','Abbreviation','rank'],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
cat_coln = df.select_dtypes(include='object').columns
```

```python
cat_coln
```

```python
num_coln = df.select_dtypes(include=['int64','float64']).columns
```

```python
num_coln
```

```python
# Handling missing values in the dataframe
NaNnumerical = df.select_dtypes(include = ['float64', 'int64']).columns
NaNcategorical = df.select_dtypes(include = ['object']).columns
df[NaNnumerical] = df[NaNnumerical].fillna(0)
df[NaNcategorical] = df[NaNcategorical].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Modifying the column names in the pandas dataframe
df.columns = df.columns.str.replace('_', ' ').str.title()
df.columns = df.columns.str.replace(' ','')
df.columns
```

```python
df.info()
```

```python
#converting datatype to a specific  columns
df = df.astype({
    'VideoViews' : 'int64',
    'VideoViewsForTheLast30Days' : 'int64',
    'SubscribersForLast30Days':'int64',
    'CreatedYear': 'int64',
    'Population': 'int64',
    'UrbanPopulation': 'int64',
    'LowestMonthlyEarnings' : 'int64',
    'HighestMonthlyEarnings' : 'int64',
    'HighestYearlyEarnings' : 'int64',
    'HighestYearlyEarnings' : 'int64'
})

```

```python
df.info()
```

```python
# exclude rows where the value in the 'VideoViews' column is equal to 0
df = df[df['VideoViews'] != 0]
df.head()
```

```python
# filter columns values equal to 0 and 1970 in the createdYvear column
df[(df['CreatedYear']==0) | (df['CreatedYear']==1970)]
print(df['CreatedYear'])
```

```python
# excluding the rows where the 'CreatedYear' column is either 0 or 1970
df = df[df['CreatedYear'] != 1970]
df = df[df['CreatedYear'] != 0]
df.head()
```

```python
import re
```

```python
# Keeping only alphanumeric characters, spaces, and specific punctuation in the 'Youtuber' and 'Title' columns of the DataFrame
pattern = re.compile(r'[^a-zA-Z0-9 .,]')
df['Youtuber'] = df['Youtuber'].replace(pattern, "", regex = True)
df['Title'] = df['Title'].replace(pattern, "", regex = True)
```

```python
df.sample(2)
```

```python
#adding  new column on the dataset ( AverageYearlyEarnings)
df['AverageYearlyEarnings'] = (df['LowestYearlyEarnings']+ df['HighestYearlyEarnings'])/2
```

```python
df.head()
```

```python
df = df.sort_values(by='CreatedMonth',ascending=True)
```

```python
df.head()
```

```python
#adding  new column on the dataset ( AverageYearlyEarnings)
df['AverageMonthlyEarnings'] = (df['LowestMonthlyEarnings']+ df['HighestMonthlyEarnings'])/2
```

```python
# Sort the DataFrame by the 'subscribers' column in descending order and select the top 10
top_10_youtubers_by_subscribers = df[['Youtuber', 'Subscribers']].sort_values(by='Subscribers', ascending=False).head(10)

# Display the top 10 YouTubers with only 'youtuber' and 'subscribers' columns
top_10_youtubers_by_subscribers
```
